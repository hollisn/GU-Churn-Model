#### Reading in the data and feature engineering


```python
import pandas as pd

# read in the data
df = pd.read_csv('data/churn.csv')

# drop redundant columns, area code is plenty
df = df.drop(['State', 'Phone'], axis=1)

# covert Area Code to string
df['Area Code'] = df['Area Code'].astype(str)

# binariaze columns
df = pd.get_dummies(df)

# drop redundant columns
df = df.drop([
              'Int\'l Plan_no', 'VMail Plan_no', 'Churn?_False.', 'Area Code_510'
             ], axis=1)

# take a look 
df.head(5)
```

#### Explore the data 

```python
import matplotlib.pyplot as plt
import seaborn
% matplotlib inline

pd.scatter_matrix(df, figsize=(20, 20), edgecolor='none', alpha=1);
```

```python
# drop the corellated features 

df_drop = df.drop(['Day Charge', 'Eve Charge', 'Night Charge', 'Intl Charge',
              'VMail Plan_yes'], axis=1)

pd.scatter_matrix(df_drop, figsize=(20, 20), edgecolor='none', alpha=1);
```

```python
print "Features remaining:", len(df_drop.columns)
```

```python
import statsmodels.api as sm

sm.graphics.plot_corr(df_drop.corr(), xnames=df_drop.columns.tolist())
plt.show();
```

Still have correlation in Area Code 415, however predictiveness dropped when it was removed, so it stays.

####  Set up test/train splits, scale

```python
# partial scale of just the continuous variables 

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_scale = pd.DataFrame()

df_scale = pd.DataFrame(scaler.fit_transform(df_drop[df_drop.columns[:11]]), columns=df_drop.columns[:11])

df_scale[df_drop.columns[11:]] = df_drop[df_drop.columns[11:]]

df_scale.to_pickle('data/dataframe_scaled')

df_scale.sample(5)
```

```python

# define X and y 
y = df_scale['Churn?_True.'].values
X = df_scale[[_ for _ in df_scale.columns if _ != 'Churn?_True.']]

# full dataset with dropped features and unbalanced classes 
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, 
                                                   random_state=42)

# duplicate X_train, y_train, before they get their classes balanced
X_train_unbal, y_train_unbal = X_train, y_train
```

#### Balancing the classes 

Sklearn defaults to stratified K-fold, preserving the distribution of classes in the data set, however in this case we do not want that we want balanced classes for training and unbalanced classes for test.

```python
print 'Total number of churn:      ', int(sum(y_train))
print 'Total number of customers:  ', len(y_train)
print 'Always negative classifier: ', 1 - (sum(y_train)/len(y_train)).round(3)
```

```python
print len(X_train), "rows of unbalanced X_train/y_train"

# put the train and test back togeather 
train = pd.DataFrame(X_train, columns=[_ for _ in df_drop.columns if _ != 'Churn?_True.'])
train['Churn?_True.'] = y_train

# dataset of just the customers who churned 
df_allchurn = train[train['Churn?_True.'] == 1]

# dataset of random sample of customers who did not churn 
df_nochurn = train[train['Churn?_True.'] == 0].sample(sum(y_train), random_state=42)

# datasets combined
df_balanced = pd.concat([df_allchurn, df_nochurn])

print df_balanced.count()[0], "rows of balanced X_train/y_train"

# split them back up again 
X_train = df_balanced[[_ for _ in df_drop.columns if _ != 'Churn?_True.']]
y_train = df_balanced['Churn?_True.'].values

print X_train.shape, "X_train shape"
print y_train.shape, "y_train shape"
```

#### Declare baseline classifier accuracy

If a classifier is perfectly negative, how high wil accuracy be?

```python
print 'Total number of churn:      ', int(sum(y_train))
print 'Total number of customers:  ', len(y_train)
print 'Always negative classifier: ', 1 - (sum(y_train)/len(y_train)).round(3)
```

#### Define a  cross-valation function

```python
from sklearn.cross_validation import cross_val_score

def xval_score(classifier, train=X_train, goal=y_train, n=5):
    
    auc = cross_val_score(classifier, train, goal,
                           cv=n, scoring='roc_auc')
    
    rec = cross_val_score(classifier, train, goal,
                           cv=n, scoring='recall') 
    
    acc = cross_val_score(classifier, train, goal,
                           cv=n, scoring='accuracy')
    
    pre = cross_val_score(classifier, train, goal,
                           cv=n, scoring='precision') 
    
    print "     __________________________________"
    print "    | AUC:", auc.mean().round(4), "| Recall:   ", rec.mean().round(4), ' |'
    print "    | Acc:", acc.mean().round(4), "| Precision:", pre.mean().round(4), ' |'
    print "     "
```

## Logistic Regression

```python
from sklearn import linear_model

lr = linear_model.LogisticRegression(penalty= 'l2')

print "Logistic Regression with ridge regularization:"
%time xval_score(lr)
```

## Extremely Random Forest

```python
from sklearn.ensemble import ExtraTreesClassifier

rf = ExtraTreesClassifier(n_estimators=40, min_samples_leaf=2, 
                          n_jobs=-1, random_state=42)

print "Vanilla Extra Random Forest:"
%time xval_score(rf, X_train, y_train)
```

#### Grid search for better parameters 

```python
param_grid = { 
    'n_estimators': [10, 20, 40, 80, 100, 120],
    'min_samples_split' : [2, 3, 4, 5],
    'min_samples_leaf': [1, 2, 3],
    'max_depth': [None],
    'bootstrap': [True, False],
    'max_features': [13, 12, 10, 4, None] }
```

Paying special attention to the scoring parameter. It has a strong affect on how the forest is tuned.  

http://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter

```python
# area under the curve grid search

from sklearn.grid_search import GridSearchCV

gs_rf_auc = GridSearchCV(estimator=rf, 
                     param_grid=param_grid, 
                     cv= 5,
                     verbose=False,
                     scoring='roc_auc',
                     n_jobs=-1)

gs_rf_auc.fit(X_train, y_train)

print
print "Best parameters" 
print gs_rf_auc.best_params_ 
print 
print "With AUC tuning:"
xval_score(gs_rf_auc.best_estimator_)
```

```python
# recall grid search

gs_rf_rec = GridSearchCV(estimator=rf, 
                     param_grid=param_grid, 
                     cv= 5,
                     verbose=False,
                     scoring='recall',
                     n_jobs=-1)

gs_rf_rec.fit(X_train, y_train)

print
print "Best parameters" 
print gs_rf_rec.best_params_ 
print 
print "With unbalanced classes and recall tuning:"
xval_score(gs_rf_rec.best_estimator_)
```

## Gradient Boosted Trees

```python
# best estimator from extreme forest 

gs_rf_rec.best_estimator_
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(loss='deviance', learning_rate=0.1,
                             n_estimators=100, subsample=0.3,
                             min_samples_split=4,
                             min_samples_leaf=1,
                             max_depth=None,
                             random_state=42,
                             max_features=13,
                             verbose=0)

xval_score(gb)                    
```

```python
# gradient boosted grid search 

param_grid = { 
    'n_estimators': [5, 8, 10, 15, 18, 20, 30, 40],
    'min_samples_split' : [2, 3, 4, 5,],
    'min_samples_leaf': [1, 3, 5, 5],
    'max_depth': [None],
    'max_features': [14, 13, 10, None],
    'learning_rate': [.03, .05, .1],
    'subsample': [.01, .03, .05, .07, .09],
    'loss': ['deviance', 'exponential'] }
```

```python
gs_bf_auc = GridSearchCV(estimator=gb, 
                     param_grid=param_grid, 
                     cv= 5,
                     verbose=False,
                     scoring='roc_auc',
                     n_jobs=-1)

gs_bf_auc.fit(X_train, y_train)

print
print "Best parameters" 
print gs_bf_auc.best_params_ 
print 
print 'Scores with balanced classes:'
xval_score(gs_bf_auc.best_estimator_)
```

```python
# balanced classes boosted grid search 

gs_bf_f1 = GridSearchCV(estimator=gb, 
                     param_grid=param_grid, 
                     cv= 5,
                     verbose=False,
                     scoring='f1',
                     n_jobs=-1)

gs_bf_f1.fit(X_train, y_train)

print
print "Best parameters" 
print gs_bf_f1.best_params_ 
print 
print 'Scores with balanced classes:'
xval_score(gs_bf_f1.best_estimator_)
```

```python
# test against the balanced test set
import sklearn.metrics as sk

models = [lr,
          gs_rf_auc.best_estimator_,
          gs_rf_rec.best_estimator_,
          gs_bf_auc.best_estimator_,
          gs_bf_f1.best_estimator_,
          ada,
          sv,
          gs_svm.best_estimator_]

names = ['Logistic Regression', 
         'GS RF AUC Tuning', 
         'GS RF Recall Tuning',
         'GS RF Gradient Boosted AUC', 
         'GS RF Gradient Boosted F1',
         'AdaBoost',
         'SVM rbf',
         'GS SVM']

cols = ['model', 'name', 'auc', 'f1', 'acc', 
        'precision', 'recall', 'tp', 'tn', 'fp', 'fn']

results_unbal = pd.DataFrame(columns=cols)

for i in xrange(len(models)):
    
    models[i].fit( X_train_unbal, y_train_unbal )
    
    auc = sk.roc_auc_score(y_test, models[i].predict(X_test)).round(3)
    f1  = sk.f1_score(y_test, models[i].predict(X_test)).round(3)
    acc = models[i].score(X_test, y_test).round(3)
    pre = sk.precision_score(y_test, models[i].predict(X_test)).round(3)
    rec = sk.recall_score(y_test, models[i].predict(X_test)).round(3)
    
    a = sk.confusion_matrix(y_test, models[i].predict(X_test))
    tp, tn, fp, fn = a[0][0], a[1][1], a[0][1], a[1][0]
    
    results_unbal.loc[len(results_unbal)]=[models[i], names[i], auc, 
                                           f1, acc, pre, rec, tp, tn, fp, fn] 
    
    
    
results_unbal.to_pickle('data/pickle_take2_results_UUU')
  

# display table, filter out the raw models 
print
print "Unbalanced train, unbalanced test"
results_unbal[[_ for _ in results_unbal.columns if _ != 'model']] #.style.highlight_max(axis=0)
```
