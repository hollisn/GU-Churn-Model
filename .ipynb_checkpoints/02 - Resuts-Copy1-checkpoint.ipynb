### Interpretation

```python
# read in the dataframe from pt 1
df_scale = pd.read_pickle('dataframe_scaled')

# define X and y 
y = df_scale['Churn?_True.'].values
X = df_scale[[_ for _ in df_scale.columns if _ != 'Churn?_True.']]

# full dataset with dropped features and unbalanced classes 
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, 
                                                   random_state=42)

df_scale.sample(5)
```

```python
clf.fit(X_train, y_train)

print sk.classification_report(y_test, clf.predict(X_test))
```

```python
# plot num features against recall

from sklearn.ensemble import ExtraTreesClassifier
from sklearn.cross_validation import cross_val_score
import matplotlib.pyplot as plt
import seaborn
% matplotlib inline


n, acc = 1, [[],[]]

while n <= 1000:
    et = ExtraTreesClassifier(bootstrap=True, class_weight=None, criterion='gini',
           max_depth=None, max_features=13, max_leaf_nodes=None,
           min_samples_leaf=1, min_samples_split=4,
           min_weight_fraction_leaf=0.0, n_estimators=n, n_jobs=-1,
           oob_score=False, random_state=42, verbose=0, warm_start=False)
    
    xacc = cross_val_score(et, X_train, y_train,
                          cv=5, scoring='recall') 
    acc[0] += [n]
    acc[1] += [xacc.mean()]
    
    if n < 10:
        n += 1
    if n < 25 and n >= 10:
        n += 3        
    if n < 150 and n >= 25:
        n += 20
    if n >= 150:
        n += 40
    
plt.figure()
plt.title("Recall VS Num Features")
plt.scatter(acc[0], acc[1])
plt.show();
```

```python
import sklearn.metrics as sk
import numpy as np


label = ['No Churn', 'Churn']

cm = sk.confusion_matrix(y_test, clf.predict(X_test), labels=None)


# scale the values, due to unbalanced classes
cm = [[1.0 * cm[0,0]/sum(cm[0]), 1.0 * cm[0,1]/sum(cm[0])], 
      [1.0 * cm[1,0]/sum(cm[1]), 1.0 * cm[1,1]/sum(cm[1])]]


# plot code adapted from http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
def plot_confusion_matrix(cm, title='Logistic Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    tick_marks = np.arange(len(label))
    plt.xticks(tick_marks, label, rotation=45)
    plt.yticks(tick_marks, label)    
    plt.title(title)
    plt.ylabel('True label')
    plt.xlabel('Predicted label');

print plot_confusion_matrix(cm)   
```

```python
# feature importances

import numpy as np
import sklearn.metrics as sk

def feat_importances(classifer, num_feat = 4):
    
    importances = classifer.feature_importances_
    std = np.std([classifer.feature_importances_ for tree in classifer.estimators_],
                 axis=0)
    indices = np.argsort(importances)[::-1]

    print("Feature ranking:")
    for f in range(num_feat):
        print("%d. %s (%f)" % (f + 1, df_scale.columns[f], importances[indices[f]]))
    
    print
    plt.figure()
    plt.title("Feature importances")
    plt.bar(range(num_feat), importances[indices[:num_feat]],
           color="r", align="center")
    plt.xlim([-1, num_feat])
    plt.show()
        
feat_importances(clf, 14)
```

### Rank Customers

```python

# initilize and empty data frame
ranking = pd.DataFrame()

# prob of churn per customer
churn_prod = clf.predict_proba(X_test)[:,1]

# indicies of churn prob sorted
indices = np.argsort(churn_prod)[::-1]

# ranking of customers
ranking[['CustID', 'Prob Churn']] = pd.DataFrame(zip(indices, 
                           churn_prod[np.argsort(churn_prod)[::-1]]) )

# reread data frame
phone = pd.read_csv('data/churn.csv')

# list customer phone numbers from indicies
phone = phone.loc[indices][['Area Code', 'Phone']]

ranking[['Area Code', 'Phone']] = phone

ranking.head(20)
```
